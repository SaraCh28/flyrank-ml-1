## 2. The question: decision, action, cost of a wrong call

### Search question
Which content pages should be prioritized for manual review based on observable search and content signals?

### Decision
Determine which pages should be reviewed first for refresh, expansion, protection, pruning, or monitoring.

### Unit of analysis
One content page.

### Output
A ranked list of content pages with a refresh priority score and reason codes explaining why each page was recommended.

### Action
SEO or content editors can review the highest-ranked pages first and decide whether they should be updated, expanded, monitored, or left unchanged.

### Cost of a wrong recommendation
A false positive wastes editorial time by recommending pages that do not need attention. A false negative means an important page may continue losing search visibility because it was not reviewed. The goal is to improve prioritization rather than guarantee better rankings.
