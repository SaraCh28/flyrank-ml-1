## 3. Success Metric

The primary success metric will be **Precision@50** because the goal is to rank pages for manual review rather than classify every page perfectly.

Additional evaluation metrics include:

- Average Precision
- ROC AUC
- Recall

These metrics better reflect how useful the ranked recommendations are to content editors.

## 5. Why ML Beats a Fixed Rule

A fixed rule may recommend pages using only one or two conditions, such as content age or low CTR.

However, page performance depends on several interacting factors including impressions, average position, CTR, freshness, engagement, and content age.

Machine learning can combine these observable signals to estimate which pages deserve manual review while still allowing human reviewers to make the final decision.

The output is intended to support prioritization rather than replace editorial judgment.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

features = [
    "content_age_days",
    "days_since_last_update",
    "impressions_90d",
    "ctr",
    "avg_position"
]

print("Features that a simple rule would struggle to combine:")

df[features].head()
```
